````python
data = pd.read_csv('student_perfomance.csv')
````

1. Посчитайте среднее кол-во часов уделяемое на сон студентами

2. Сколько процентов студентов посещают внеклассовые активности?

````python
activities = (data["Extracurricular_Activities"] == "Yes").sum()
total = len(data)
percent_activities = activities / total * 100
percent_activities
````

3. Правда ли, что в среднем девушки сдают экзамены лучше, чем мужчины

````python
male = data[data["Gender"] == "Male"]
female = data[data["Gender"] == "Female"]
male_average = male[["Exam_Score", "Previous_Scores"]].mean().mean()
female_average = female[["Exam_Score", "Previous_Scores"]].mean().mean()
if female_average > male_average:
    print("Да, в среднем девушки сдают экзамены лучше, чем мужчины.")
elif female_average < male_average:
    print("Нет, в среднем мужчины сдают экзамены лучше, чем девушки.")
else:
    print("Средние оценки на экзаменах у девушек и мужчин одинаковые.")
````

4. Выведите 25 и 75 перцентили времени сна студентов

````python
sleep_25th = data["Sleep_Hours"].quantile(0.25)
sleep_75th = data["Sleep_Hours"].quantile(0.75)
print(f"25-й перцентиль времени сна: {sleep_25th:.2f} часов")
print(f"75-й перцентиль времени сна: {sleep_75th:.2f} часов")
````

5. Правда ли, что те, кто спит больше(75 перцентиль и выше) сдает экзамен лучше, чем студенты с недосыпом(25 перцентиль и меньше)

````python
high = data[data["Sleep_Hours"] >= sleep_75th]
normal = data[(data["Sleep_Hours"] < sleep_75th) & (data["Sleep_Hours"] > sleep_25th)]
low = data[data["Sleep_Hours"] <= sleep_25th]

high_average = high["Exam_Score"].mean()
normal_average = normal["Exam_Score"].mean()
low_average = low["Exam_Score"].mean()

if high_average > low_average:
    print("Да, студенты, которые спят больше, в среднем сдают экзамены лучше.")
elif high_average < low_average:
    print("Нет, студенты с недосыпом в среднем сдают экзамены лучше")
else:
    print("Средние оценки на экзаменах у студентов с разным количеством сна не различаются.")
````

6. Влияет ли физическая активность на результаты экзамена?

````python
low_activity = data[data["Physical_Activity"] <= data["Physical_Activity"].quantile(0.25)]
normal_activity = data[(data["Physical_Activity"] > data["Physical_Activity"].quantile(0.25)) & (data['Physical_Activity'] <= data['Physical_Activity'].quantile(0.75))]
high_activity = data[data["Physical_Activity"] > data["Physical_Activity"].quantile(0.75)]

low_activity_average = low_activity["Exam_Score"].mean()
normal_activity_average = normal_activity["Exam_Score"].mean()
high_activity_average = high_activity["Exam_Score"].mean()

if high_activity_average > low_activity_average:
    print("Да, студенты с высокой физической активностью в среднем сдают экзамены лучше.")
elif high_activity_average < low_activity_average:
    print("Нет, студенты с низкой физической активностью в среднем сдают экзамены лучше.")
else:
    print("Средние оценки на экзаменах у студентов с разным уровнем физической активности не различаются.")

````

7. Правда ли, что слуденты, чьи родители имеют высшее образование(College || Postgraduate), сдают экзамены лучше, чем остальные?

````python
yes_education = data[(data["Parental_Education_Level"] == "College") | (data["Parental_Education_Level"] == "Postgraduate")]
no_education = data[(data["Parental_Education_Level"] != "College") | (data["Parental_Education_Level"] != "Postgraduate")]

yes_average = yes_education[["Exam_Score", "Previous_Scores"]].mean().mean()
no_average = no_education[["Exam_Score", "Previous_Scores"]].mean().mean()

if yes_average > no_average:
    print("Да, студенты, чьи родители имеют высшее образование, в среднем сдают экзамены лучше, чем остальные.")
elif yes_average < no_average:
    print("Нет, студенты, чьи родители не имеют высшего образования, в среднем сдают экзамены лучше, чем остальные.")
else:
    print("Средние оценки на экзаменах у студентов с родителями с разным уровнем образования не различаются.")
````

8. Как влияет участие родителей на результаты экзаменов

````python

yes_active = data[data["Parental_Involvement"] == "High"][["Exam_Score", "Previous_Scores"]].mean().mean()
no_active = data[data["Parental_Involvement"] == "Low"][["Exam_Score", "Previous_Scores"]].mean().mean()
medium_active = data[data["Parental_Involvement"] == "Medium"][["Exam_Score", "Previous_Scores"]].mean().mean()

if (yes_active > no_active) & (yes_active > medium_active):
    print("Cтуденты, чьи родители активно участвуют в их обучении, в среднем сдают экзамены лучше, чем остальные.")
elif (no_active > yes_active) & (no_active > medium_active):
    print("Cтуденты, чьи родители не участвуют или участвуют мало, в среднем сдают экзамены лучше, чем остальные.")
elif (medium_active > yes_active) & (medium_active > no_active):
    print("Cтуденты, чьи родители участвуют умеренно, в среднем сдают экзамены лучше, чем остальные.")
elif (no_active == yes_active) & (no_active == medium_active):
    print("Средние оценки на экзаменах у студентов с разным уровнем участия родителей не различаются.")
````

9. Правда ли, что в семьях с доходом больше среднего дети сдают экзамены лучше?

````python
high_income = data[data["Family_Income"] == "High"]
medium_income = data[data["Family_Income"] == "Medium"]
low_income = data[data["Family_Income"] == "Low"]

high_income_average = high_income[["Exam_Score", "Previous_Scores"]].mean().mean()
low_income_average = low_income[["Exam_Score", "Previous_Scores"]].mean().mean()
medium_income_average = medium_income[["Exam_Score", "Previous_Scores"]].mean().mean()

if (high_income_average > low_income_average) & (high_income_average > medium_income_average):
    print("Да, студенты из семей с доходом выше среднего в среднем сдают экзамены лучше.")
elif (high_income_average < low_income_average) & (medium_income_average < low_income_average):
    print("Нет, студенты из семей с доходом ниже среднего в среднем сдают экзамены лучше.")
elif (medium_income_average > low_income_average) & (medium_income_average > high_income_average):
    print("Нет, студенты из семей со средним доходом в среднем сдают экзамены лучше.")
else:
    print("Средние оценки на экзаменах у студентов из семей с разным уровнем дохода не различаются.")
````
